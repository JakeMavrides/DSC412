```python
import pandas as pd
import numpy as np
```

```python
# California housing market dataset
df = pd.read_csv('housing.csv')
df.shape
```

```python
df.head()
```

```python
X = df[['longitude', 'latitude', 'median_income']] # We will just use these 3 variables
X.head()
```

```python
from sklearn.cluster import KMeans
```

```python
k = 6 # num clusters
kmeans = KMeans(n_clusters=k)
#X['Cluster'] =
kmeans.fit(X) # Create a column in the dataframe named 'cluster'
```

```python
X.head()
```

```python
centers = kmeans.cluster_centers_
```

```python
centers.shape
```

```python
y_pred = kmeans.predict(X)
print(y_pred)
```

```python
df['pred'] = y_pred # Add our predicted cluster to our dataframe
```

```python
df.head()
```

```python
# Lets visualize the clustering:
import seaborn as sns
```

```python
sns.relplot(x='longitude', y='latitude', hue='pred', data=df, height=k)
```
